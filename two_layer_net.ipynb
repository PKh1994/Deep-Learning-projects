We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:
The vanishing gradient problem arises when the gradient flow during backpropagation becomes close to zero.

The sigmoid function exhibits the vanishing gradient problem because the gradient approaches zero for very large positive and negative input values, particularly in the tail portions of the function. As an illustrative one-dimensional example, consider highly extreme values, such as [-1e3, 1e3], which can lead to saturation.

One of the significant advantages of ReLU over the sigmoid function is its resilience to the vanishing gradient problem. This resilience is attributed to its linear response to positive input values. Specifically, ReLU's gradient is either 0 for negative inputs or 1 for positive inputs. While ReLU can experience the vanishing gradient problem when all input values are negative (a scenario that is less likely), it may result in certain neurons failing to further train. This issue is commonly referred to as the "dying ReLU problem." For example, in a one-dimensional context, consider having only negative values, like [-1, -2, -3], which can lead to the vanishing gradient problem.

Leaky ReLU is designed to address the "dying" neuron problem of ReLU by introducing a slight negative slope for negative values, defined as follows: if x < 0, then 0.01x; otherwise, x. Leaky ReLU is intended to mitigate the vanishing gradient problem. However, it's important to note that the max(0.01x, x) function is not continuous at x = 0, which results in an undefined gradient at that point. If not explicitly handled in the code, a one-dimensional scenario that can lead to zero gradients is having all zero values, e.g., [0, 0, 0], a situation that typically arises due to poor network initialization
