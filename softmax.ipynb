```python
def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000, num_dev=500):
    """
    Load the CIFAR-10 dataset from disk and perform preprocessing to prepare
    it for the linear classifier. These are the same steps as we used for the
    SVM, but condensed to a single function.
    """
    # Load the raw CIFAR-10 data
    cifar10_dir = 'cse493g1/datasets/cifar-10-batches-py'

    # Cleaning up variables to prevent loading data multiple times (which may cause memory issue)
    try:
       del X_train, y_train
       del X_test, y_test
       print('Clear previously loaded data.')
    except:
       pass

    X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

    # subsample the data
    mask = list(range(num_training, num_training + num_validation))
    X_val = X_train[mask]
    y_val = y_train[mask]
    mask = list(range(num_training))
    X_train = X_train[mask]
    y_train = y_train[mask]
    mask = list(range(num_test))
    X_test = X_test[mask]
    y_test = y_test[mask]
    mask = np.random.choice(num_training, num_dev, replace=False)
    X_dev = X_train[mask]
    y_dev = y_train[mask]

    # Preprocessing: reshape the image data into rows
    X_train = np.reshape(X_train, (X_train.shape[0], -1))
    X_val = np.reshape(X_val, (X_val.shape[0], -1))
    X_test = np.reshape(X_test, (X_test.shape[0], -1))
    X_dev = np.reshape(X_dev, (X_dev.shape[0], -1))

    # Normalize the data: subtract the mean image
    mean_image = np.mean(X_train, axis = 0)
    X_train -= mean_image
    X_val -= mean_image
    X_test -= mean_image
    X_dev -= mean_image

    # add bias dimension and transform into columns
    X_train = np.hstack([X_train, np.ones((X_train.shape[0], 1))])
    X_val = np.hstack([X_val, np.ones((X_val.shape[0], 1))])
    X_test = np.hstack([X_test, np.ones((X_test.shape[0], 1))])
    X_dev = np.hstack([X_dev, np.ones((X_dev.shape[0], 1))])

    return X_train, y_train, X_val, y_val, X_test, y_test, X_dev, y_dev


# Invoke the above function to get our data.
X_train, y_train, X_val, y_val, X_test, y_test, X_dev, y_dev = get_CIFAR10_data()
print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
print('dev data shape: ', X_dev.shape)
print('dev labels shape: ', y_dev.shape)
```

**Inline Question 1**

Why do we expect our loss to be close to -log(0.1)? Explain briefly.

$\color{blue}{\textit Your Answer:}$ Because we use a randomly initialized weight matrix W, we anticipate that each class is selected at random, which means that each class is expected to have roughly the same scores, denoted as s_j, for a given input image. Given that there are 10 classes, this implies that the scores s_j and s_k should be approximately equal for all pairs of classes (j and k) in the range of 0 to 9. As a result, the exponential values e^{s_j} and e^{s_k} should also be quite similar for all classes.

Consequently, when we calculate the expectation of the term inside the logarithm function, it simplifies to e^{s_k} divided by the sum of all the exponential values e^{s_j} for j ranging from 0 to 9. This ratio can be further simplified to e^{s_j} divided by 10 times e^{s_j}, which ultimately results in 1/10.

**Inline Question 2** - *True or False*

Suppose the overall training loss is defined as the sum of the per-datapoint loss over all training examples. It is possible to add a new datapoint to a training set that would leave the SVM loss unchanged, but this is not the case with the Softmax classifier loss.

$\color{blue}{\textit Your Answer: True}$


$\color{blue}{\textit Your Explanation:}$ The SVM loss remains unaffected as long as the new data point's SVM loss is zero. This occurs when the incorrect classes meet the margin requirement, i.e., their scores are at least delta less than the score of the correct class. To illustrate, assuming (i) a delta of 1 for the SVM, (ii) the correct class is #0, and (iii) the introduction of a new data point, #3, resulting in scores [3, 1, 2]. In this scenario, the SVM loss for the new data point is 0 (preserving the overall SVM loss) because it meets the margin condition. This is calculated as max(0, 1 - 3 + 1) + max(0, 2 - 3 + 1) = 0.

In contrast, the Softmax classifier would incur a greater loss when a new data point is added because it diminishes the score of the correct class after normalization. This reduction negatively impacts the loss. The reason for this lies in the Softmax classifier's consideration of all individual scores in its loss calculation, unlike the SVM loss, which disregards the individual scores as long as the incorrect classes maintain a score difference of at least delta from the correct class. In other words, the precise magnitude of the score difference between the incorrect classes and the correct class (provided it is at least delta) is not a concern for the SVM, whereas it matters for the Softmax.
